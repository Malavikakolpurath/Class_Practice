```python
### MANAGING OUTLIERS
#Using beer-servings.csv data
```

```python
num_cols=['beer_servings','spirit_servings'	,'wine_servings',	'total_litres_of_pure_alcohol']
```

```python
# first we drow a boxplot
for i in num_cols:
  plt.figure()
  plt.boxplot(data[i])
  plt.title(i)
```

```python
# 25th percentile = 25/100 *(n+1)
 # IQR = Q3-Q1
 # lowest =Q1 -1.5*IQR
 # upper =Q3 + 1.5*IQR
 # IQR =inter quartile range
```

```python
# now we are calculating Q1 Q2 and Q3 using coding
Q1=np.percentile(data['spirit_servings'],25,method='midpoint')
Q2=np.percentile(data['spirit_servings'],50,method='midpoint')
Q3=np.percentile(data['spirit_servings'],75,method='midpoint')
```

```python
print("Q1 is",Q1)
print("Q2 is",Q2)
print("Q3 is",Q3)
```

```python
data['spirit_servings'].median()
```

```python
IQR =Q3-Q1
```

```python
low_lim=Q1-1.5*IQR
up_lim=Q3+1.5*IQR
print("min is",low_lim)
print("max is",up_lim)
```

```python
outlier=[]
for x in data['spirit_servings']:
  if((x>up_lim) or (x<low_lim)):
     outlier.append(x)
```

```python
outlier
# outlier kandu pidichu
```

```python
ind1 =data['spirit_servings']>up_lim
data.loc[ind1].index
# outlier nte index kandu pidichu
```
