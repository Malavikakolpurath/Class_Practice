```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data.columns
```

```python
data.info()
```

```python
data.shape
```

```python
data.describe()
```

```python
plt.bar(data['Category'],data['Sales'])
plt.show()
```

```python
plt.bar(data['Sub-Category'],data['Sales'])
plt.show()
```

```python
plt.bar(data['Sub-Category'],data['Sales'])
plt.xticks(rotation=90)
plt.show()
```

```python
data['Segment'].value_counts()
```

```python
plt.bar(data['Segment'],data['Sales'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(data['Segment'],data['Profit'])
plt.xticks(rotation=90)
plt.show()
```

```python
print(data['State'].unique())
```

```python
print(data['State'].nunique())
```

```python
loss_df=data[data['Profit']<0]
loss_df
```

```python
loss_df.shape
```

```python
loss_df.describe()
```

```python
Total_loss=loss_df['Profit'].sum()
Total_loss
```

```python
loss_df.groupby(by='Segment').sum()
```

```python
loss_df.groupby(by='Sub-Category').sum()
```

```python
loss_df.groupby(by='City').sum()
```

```python
loss_df.groupby(by='City').sum().sort_values('Profit',ascending=True).head(10)
```
