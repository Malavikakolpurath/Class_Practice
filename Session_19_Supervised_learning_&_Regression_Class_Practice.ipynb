```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('tip-amount.xlsx')
data
```

```python
## now we make the machine learn the data and predict the data
```

```python

### now define what is x and what is y
y=data['Observed tip amount(yi)']
```

```python
x=data.drop(['Observed tip amount(yi)'],axis=1)
```

```python
y
```

```python
##now i want to split data. for that,
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
```

```python
### to fit the model,
from sklearn import linear_model
lr=linear_model.LinearRegression()
```

```python
## now we create model (means to learn from x train and y train )
model=lr.fit(x_train,y_train)
```

```python
### predict cheyaan parayunu(exam)
predictions=model.predict(x_test)
```

```python
## to find error
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,predictions)
```

```python
## r2 score is a statistical measure that represent goodness of fit
from sklearn.metrics import r2_score
r2_score(y_test,predictions)
```
