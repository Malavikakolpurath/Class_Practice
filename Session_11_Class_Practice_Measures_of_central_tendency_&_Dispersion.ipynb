```python
import numpy as np
```

```python
### mean , median, and mode
```

```python
speed=[99,86,87,88,111,86,103,87,94,78,77,85,86]
speed_mean=np.mean(speed)
print(speed_mean)
```

```python
speed_median=np.median(speed)
print(speed_median)
```

```python
from scipy import stats
speed_mode=stats.mode(speed)
print(speed_mode)
```

```python
import pandas as pd
df=pd.DataFrame({'Player':['A','B','C','D','E','F','G','H'],
                 'game1':[18,22,19,14,14,11,20,25],
                 'game2':[5,7,7,9,12,9,9,4],
                 'game3':[11,8,10,6,6,5,8,12],
                 'game4':[9,8,10,9,14,15,10,11]})
df
```

```python
# mean for each column
game_mean=df.mean(numeric_only=True)
print(game_mean)
```

```python
game_median=df.median(numeric_only=True)
print(game_median)
```

```python
game_mode=df.mode(numeric_only=True)
print(game_mode)
```
