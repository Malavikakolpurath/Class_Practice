```python
import random as random
students=['jane','joe','beck','tom','liam','culhane','candane','ana','mari','steffi','adam']
marks=[]
for i in range(0,len(students)):
    marks.append(random.randint(0,101))

plt.xlabel('students')
plt.ylabel('marks')
plt.title('CLASS RECORDS')

plt.plot(students,marks,color="k",
             linestyle="dashed",marker="o",
             markerfacecolor="c",markersize=8,markeredgecolor="r")
```

```python
### new data set for pie chart (pie chart for 100% of something hole percentage lu oro particular nu ethra percentage)

cars =['audi','bmw','ford','tesla','jaguar','mercedes']

nums=[23,17,35,29,12,41]

# creating plot

fig=plt.figure(figsize=(10,7))
plt.pie(nums,labels=cars)
plt.title('car sold')

#show plot
plt.show()
```

```python
cars =['audi','bmw','ford','tesla','jaguar','mercedes']

nums=[23,17,35,29,12,41]
explode=[0.1,0,0,0,0.1,0]
# creating plot

fig=plt.figure(figsize=(10,7))
plt.pie(nums,labels=cars,explode=explode)
plt.title('car sold')

#show plot
plt.show()
```

```python
cars =['audi','bmw','ford','tesla','jaguar','mercedes']

nums=[23,17,35,29,12,41]

# creating plot

fig=plt.figure(figsize=(10,7))
plt.pie(nums,labels=cars,autopct='%1.1f%%')
plt.title('car sold')

#show plot
plt.show()
```

```python
cars =['audi','bmw','ford','tesla','jaguar','mercedes']

nums=[23,17,35,29,12,41]

# creating plot

fig=plt.figure(figsize=(5,4))
plt.pie(nums,labels=cars)
plt.title('car sold')

#show plot
plt.show()
```

```python
x=[1,2,3,4]
y=[4,1,3,6]

fig,ax=plt.subplots(1,2,sharey=True)
## sharey=True chart ne oru paper lu thanne share cheyaan
ax[0].plot(x,y)
ax[0].set_title('Line chart')

x1=[5,6,7,8]
y1=[1,3,5,2]

ax[1].scatter(x1,y1)
ax[1].set_title('scatter chart')
```
