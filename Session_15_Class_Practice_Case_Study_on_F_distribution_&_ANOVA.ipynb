```python
import scipy.stats as stats
tech_a=[85,88,84,90,92]
tech_b=[78,82,80,85,88]
tech_c=[92,95,89,96,91]
f_stat,p_val=stats.f_oneway(tech_a,tech_b,tech_c)
alpha=0.05
print('F_stat: ',f_stat)
print('P_val: ',p_val)
```

```python
if p_val<alpha:
  print('Reject Null Hypothesis :There is significance differance in mean scores among study techniques')
else:
  print('Cannot Reject Null Hypothesis :There is no significance differance in mean scores among study techniques')
```

```python
import numpy as np
import pandas as pd
df=pd.DataFrame({'water':np.repeat(['daily','weekly'],15),
                 'sun':np.tile(np.repeat(['low','med','high'],5),2),
                 'height':[6,6,6,5,6,5,5,6,4,5,
                           6,6,7,8,7,3,4,4,4,5,
                           4,4,4,4,4,5,6,6,7,8,]})
```

```python
df
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
model=ols('height~C(water)+C(sun)+C(water):C(sun)',data=df).fit()
sm.stats.anova_lm(model,typ=2)
```

```python
data=pd.DataFrame({'AgeGroup':['Age Grp 1']*15+['Age Grp 2']*15,
                   'Technique':['Tech A']*5+['Tech B']*5+['Tech C']*5+
                                ['Tech A']*5+['Tech B']*5+['Tech C']*5,
                   'score':[85,88,84,90,92,78,82,80,85,88,92,95,89,96,91,
                            78,80,82,85,89,92,95,91,96,90,84,88,85,90,82]})
data
```

```python
formula='score~C(AgeGroup)+C(Technique)+C(AgeGroup):C(Technique)'
model=ols(formula,data=data).fit()
anova_table=sm.stats.anova_lm(model,typ=2)
print(anova_table)
```

```python
alpha = 0.05
if any(anova_table['PR(>F)']<alpha):
  print('There is significant interaction between factors')

  age_grp =anova_table.loc['C(AgeGroup)','PR(>F)']<alpha
  tech=anova_table.loc['C(Technique)','PR(>F)']<alpha

  if age_grp:
    print('Age grp has significant effect')
  else:
     print('Age grp has no significant effect')
  if tech:
     print('Technique has significant effect')
  else:
    print('Technique has no significant effect')
else:
  print('No significant interaction between factors')
```
